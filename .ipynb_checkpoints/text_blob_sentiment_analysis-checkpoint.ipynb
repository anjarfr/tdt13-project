# Natural language processing

```python
import nltk
nltk.download('stopwords')
```

Convert to lowercase, remove punctuation and tokenize

```python
import string

punct = string.punctuation + '“”‘’'

def clean(article):
    lowercase = article.lower()
    punctuation = "".join([char for char in lowercase if char not in punct])
    return punctuation
```

Remove stop words

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')

def remove_stop_words(tokens):
    filtered = [word for word in tokens if word not in stop_words]
    return filtered
```

Define NLP pre_processing function

```python
def pre_process(x):
    c = clean(x)
    f = remove_stop_words(c)
    return f
```

```python
data['cleaned'] = data['text'].apply(pre_process)
data
```

# Sentiment analysis

```python
from nltk.corpus import sentiwordnet as swn
```

To perform sentiment analysis, we must first POS tag the words in the corpus, and make them compatible with SentiWordNet

```python
from nltk.corpus import wordnet

def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return ''
```

```python
def pos_tag(x):
    p_t = nltk.pos_tag(x)
    swn_t = list(map(lambda word: (word[0], get_wordnet_pos(word[1])), p_t))
    return swn_t
```

```python
data['tagged'] = data['cleaned'].apply(pos_tag)
data
```

Use these to calculate the score of each document

```python
def calc_score(doc):
    pos_score = 0
    neg_score = 0
    obj_score = 0
    word_count = len(doc)
    for word in doc:
        s = list(swn.senti_synsets(word[0], word[1]))
        if not s:
            continue
        pos_score += s[0].pos_score()
        neg_score += s[0].neg_score()
        obj_score += s[0].obj_score()
    if word_count:
        pos, neg, obj = pos_score/word_count, neg_score/word_count, obj_score/word_count
    else:
        pos, neg, obj = 0, 0, 0
    return pd.Series([pos, neg, obj])
```

```python
data[['pos_score', 'neg_score', 'obj_score']] = data['tagged'].apply(calc_score)
```

Normalize the scores

```python
from sklearn import preprocessing

scaler = preprocessing.MinMaxScaler(feature_range=(0, 1))
pos_scores = data[['pos_score']].values.astype(float)
neg_scores = data[['neg_score']].values.astype(float)
obj_scores = data[['obj_score']].values.astype(float)
```

```python
data[['norm_pos_score']] = scaler.fit_transform(data[['pos_score']])
data[['norm_neg_score']] = scaler.fit_transform(data[['neg_score']])
data[['norm_obj_score']] = scaler.fit_transform(data[['obj_score']])
```

# Weak label using snorkel

```python
data = pd.read_csv('sentiment_scored_data.csv')
```

```python
ABSTAIN = -1
REAL = 0
FAKE = 1
```

```python
from snorkel.labeling import labeling_function, LabelingFunction

@labeling_function()
def lf_is_negative(x):
    # Return a label of FAKE if sentiment is negative, otherwise ABSTAIN
    return FAKE if x.norm_neg_score >= 0.065 else ABSTAIN

@labeling_function()
def lf_is_positive(x):
    # Return a label of FAKE if sentiment is positive, otherwise ABSTAIN
    return FAKE if x.norm_pos_score >= 0.1 else ABSTAIN

@labeling_function()
def lf_is_objective(x):
    # Return a label of REAL if sentiment is neutral, otherwise ABSTAIN
    return REAL if x.norm_obj_score > 0.9 else ABSTAIN

@labeling_function()
def lf_is_objective(x):
    # Return a label of REAL if sentiment is neutral, otherwise ABSTAIN
    return REAL if x.norm_obj_score > 0.9 else ABSTAIN

def keyword_lookup(x, keywords, label):
    if any(word in x.cleaned for word in keywords):
        return label
    return ABSTAIN

def make_keyword_lf(keywords, label=FAKE):
    return LabelingFunction(
        name=f"keyword_{keywords[0]}",
        f=keyword_lookup,
        resources=dict(keywords=keywords, label=label),
    )

keyword_reuters = make_keyword_lf(keywords=["reuters"], label=REAL)
```

```python
from snorkel.labeling import PandasLFApplier

lfs = [lf_is_negative, lf_is_positive, lf_is_objective, keyword_reuters]

applier = PandasLFApplier(lfs=lfs)
L_train = applier.apply(df=data)
```

```python
from snorkel.labeling import LFAnalysis

LFAnalysis(L=L_train, lfs=lfs).lf_summary()
```

```python
from snorkel.analysis import get_label_buckets

buckets = get_label_buckets(L_train[:, 0], L_train[:, 1], L_train[:, 2])
```

```python
from snorkel.labeling.model import MajorityLabelVoter

majority_model = MajorityLabelVoter()
preds_train = majority_model.predict(L=L_train)
```

# Word embeddings

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
```

```python
docs = data.text
```

```python
t = Tokenizer()
t.fit_on_texts(docs)
```

```python
vocab_size = len(t.word_index) + 1
vocab_size
```

```python
encoded_docs = t.texts_to_sequences(docs)
max_length = len(max(encoded_docs, key=len))
```

```python
padded_docs = pad_sequences(encoded_docs, maxlen=max_length, padding='post')
len(padded_docs)
```

```python
embeddings_index = dict()
with open('glove.6B/glove.6B.100d.txt') as f:
    for line in f:
        values = line.split()
        word = values[0]
        coefs = np.asarray(values[1:], dtype='float32')
        embeddings_index[word] = coefs
```

```python
print('Loaded %s word vectors.' % len(embeddings_index))
```

```python
embedding_matrix = np.zeros((vocab_size, 100))
for word, i in t.word_index.items():
    embedding_vector = embeddings_index.get(word)
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector
```

```python
embedding_matrix.shape
```

# Split data

```python
from sklearn.model_selection import train_test_split

X = pd.DataFrame(padded_docs)
y = data[['weak_label', 'label']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify=y)
```

```python
index_names = y_train[y_train['weak_label'] == -1].index 
```

# Create model

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers import Embedding
```

```python
model = Sequential()
e = Embedding(vocab_size, 100, weights=[embedding_matrix], input_length=max_length, trainable=False)
model.add(e)
model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))
# compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
# summarize the model
print(model.summary())
```

```python
model.fit(X_train.values, y_train['weak_label'].values, epochs=10, verbose=1, batch_size=16, validation_split=0.1)
```

```python
model.save('snorkel_keras_model')
```

```python
loss, accuracy = model.evaluate(X_test, y_test['label'].values, verbose=1)
print('Accuracy: %f' % (accuracy*100))
```
